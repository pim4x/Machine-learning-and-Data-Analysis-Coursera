```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

В приложенном файле — данные по ежемесячному уровню производства молока в фунтах на одну корову:

```python
milk = pd.read_csv('monthly-milk-production.csv',';', index_col=['month'], parse_dates=['month'], dayfirst=True)
```

```python
milk.columns
```

Постройте его график:

```python
plt.plot(milk['milk'])
```

Примените к ряду из предыдущего вопроса критерий Дики-Фуллера.

Чему равен полученный достигаемый уровень значимости? Округлите до двух знаков после десятичной точки.

```python
import statsmodels.api as sm
Dickey_Fuller_crit = sm.tsa.stattools.adfuller(milk['milk'])
```

```python
print(round(Dickey_Fuller_crit[1],2))
```

Часто, когда вы имеете дело с величинами, представляющими собой сумму значений показателя за каждый день или за каждый рабочий день, имеет смысл перед началом прогнозирования поделить весь ряд на число дней в периоде. Например, если поделить ряд с объёмом производства молока на одну корову на число дней в месяце, полученная величина будет меняться более плавно, и для неё легче будет построить прогнозирующую модель.

Используйте число дней в месяце для того, чтобы вычислить новый показатель — среднее дневное число полученного молока на одну корову. Постройте график этого ряда и убедитесь, что он стал более гладким.

В качестве ответа введите сумму значений полученного показателя за весь период; округлите до двух знаков после десятичной точки.

```python
milk_day_average = milk.apply(lambda x: x / x.index.days_in_month, axis=0)
```

```python
plt.plot(milk_day_average['milk'])
```

```python
round(sum(milk_day_average['milk']), 2)
```

Для ряда со средним дневным количеством молока на корову из предыдущего вопроса давайте с помощью критерия Дики-Фуллера подберём порядок дифференцирования, при котором ряд становится стационарным. 

Первое дифференцирование сезонное:

```python
d1 = (milk_day_average - milk_day_average.shift(12)).dropna()
Dickey_Fuller_crit = sm.tsa.stattools.adfuller(d1['milk'])
print(round(Dickey_Fuller_crit[1],2))
```

Второе дифференцирование обычное:

```python
d2 = (d1 - d1.shift(1)).dropna()
Dickey_Fuller_crit = sm.tsa.stattools.adfuller(d2['milk'])
print(Dickey_Fuller_crit[1])
```

Для стационарного ряда из предыдущего вопроса (продифференцированного столько раз, сколько вы посчитали нужным) постройте график автокорреляционной функции. Исходя из этого графика, какое начальное приближение вы предложили бы для параметра QQQ в модели SARIMA?

```python
sm.graphics.tsa.plot_acf(d2.values.squeeze(), lags=50)
```
